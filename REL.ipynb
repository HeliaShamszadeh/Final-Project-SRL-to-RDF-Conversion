```python
# Set up paths (assuming you've uploaded or cloned the repository)
import sys
sys.path.append("/content/REL/src")

# Test imports
from REL.mention_detection import MentionDetection
from REL.entity_disambiguation import EntityDisambiguation
from REL.ner import load_flair_ner

print("REL imported successfully.")
```

```python
import os

base_url = "/content/REL/data"
wiki_version = "wiki_2014"

mention_detection = MentionDetection(base_url, wiki_version)
tagger_ner = load_flair_ner("ner-fast")

long_text = """Rezazadeh was born in Iran and was a wrestler. Barack Obama, the 44th President of the United States, was born in Honolulu, Hawaii.
He served two terms in office from 2009 to 2017.
During his presidency, Obama signed the Affordable Care Act into law,
a significant reform of the healthcare system. In 2015, Obama visited Cuba,
marking the first time a sitting U.S. president had visited the island in nearly 90 years.
After leaving office, he continued his work on global issues and humanitarian causes.
Rouhani was the president of IRIB.
Jordan moved to Jordan last summer and started working at Apple.
While there, he often drank apple juice in the office cafeteria,
but his favorite project remained a demo called Mercury — he wasn’t sure
if the name came from the Roman god, the planet, or the car model.
In Washington, he gave a keynote alongside Washington Post editors,
and later visited the Washington Monument.
His next stop was Paris, where he attended the Paris Agreement summit
and enjoyed watching Paris Hilton’s reality show.
Finally, he ended the trip in Cambridge to collaborate on a new Python library.
After leaving Oracle, Taylor joined Amazon’s company leadership team to study the Amazon River basin.
Next, she drove her Jaguar down the Pacific Coast Highway, listening to music by The Doors.
In the spring, she visited Turkey and sampled fresh orange juice in Orange County, then flew
to Java to give a talk on Java programming. On her way home, she stopped in
Georgia for a Georgia O’Keeffe exhibit—and later picked up a pet python on her way through
the Python Institute’s campus.
"""

text_input = {
    "doc1": [long_text, []]
}

mentions_dataset, n_mentions = mention_detection.find_mentions(text_input, tagger_ner)

config = {
    "mode": "eval",
    "model_path": "ed-wiki-2014"
}
model = EntityDisambiguation(base_url, wiki_version, config)

predictions, _ = model.predict(mentions_dataset)


# Display results
for doc_id, entities in predictions.items():
    print(f"\nDocument: {doc_id}")
    for entity in entities:
        mention = entity['mention']
        prediction = entity['prediction']
        wiki_url = f"https://en.wikipedia.org/wiki/{prediction}"

        print(f"🔸 Mention: {mention}")
        print(f"🔗 Prediction: {prediction}")
        print(f"🌐 Link: {wiki_url}")
```
